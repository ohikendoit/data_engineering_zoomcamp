```python
import psycopg2
import pandas as pd
pd.__version__
```

```python
#read the original parquet format file
df = pd.read_parquet('yellow_tripdata_2022-01.parquet')
```

```python
#conver the parquet format file into csv
pd.read_parquet('yellow_tripdata_2022-01.parquet').to_csv('yellow_tripdata_2022-01.csv')
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
#sqlalchemy - open source sql kit
from sqlalchemy import create_engine

engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
#Iteration to chunk the big files to smaller sizes

df_iter = pd.read_csv('yellow_tripdata_2022-01.csv', iterator=True, chunksize=100000)
```

```python
df = next(df_iter) #see the list of smaller files
df
```

```python
len(df)
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

```python
from time import time

while True:
    t_start = time()
    
    df = next(df_iter)
    
    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
    
    df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
    
    t_end = time()
    
    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```
